```python
model = HLinUCB(alpha_a=0.1, alpha_u=0.1, lam1=1, lam2=1, context_columns=context_columns, 
                user_column='user_id', decisions_column='book_id')
model.fit(train_data, train_data['binary_rating'])
```

```python
raw_predictions[0]
```

```python
def select_top_n(probabilities, n):
        predictions = []
        for user in probabilities:
            top_n = []
            for _ in range(n):
                if not user:
                    continue
                    
                prediction = max(user, key=user.get)
                top_n.append(prediction)
                user.pop(prediction)
                

            predictions.append(top_n)

        return predictions
```

```python
import copy
top_n_input = copy.deepcopy(raw_predictions)
top_n = select_top_n(top_n_input, n=10)
```

```python
pickle.dump(top_n, open('data/backups/hlinucb_top10.pkl', 'wb'))
```

```python
max_p = evaluate_df['pred_proba'].max()
max_p
```

```python
# Ref: https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html
def scale_prediction(x, min_p, max_p):
    return np.rint((x - min_p) / (max_p - min_p))

evaluate_df['prediction'] = evaluate_df.apply(lambda x: scale_prediction(x['pred_proba'], min_p, max_p), axis=1)

```
